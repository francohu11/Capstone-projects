```python
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('per_game_stats.csv')
```

```python
df.dtypes
```

```python
subset = df.loc[:, ['Player','AST', 'MP']]
subset['AST/MP'] = subset['AST'] / subset['MP']
```

```python
subset_MJ = subset.loc[subset['Player'] == 'Michael Jordan',['AST/MP'] ]
subset_LBJ = subset.loc[subset['Player'] == 'Lebron James',['AST/MP'] ]
subset_kobe = subset.loc[subset['Player'] == 'Kobe Bryant',['AST/MP'] ]

plt.figure(figsize = (12,8))
sns.distplot(subset_MJ, label='MJ')
sns.distplot(subset_LBJ, label='LBJ')
sns.distplot(subset_kobe, label='kobe')
plt.legend()
```
